## üéØ 1. MAE (–°—Ä–µ–¥–Ω—è—è –∞–±—Å–æ–ª—é—Ç–Ω–∞—è –æ—à–∏–±–∫–∞, Mean Absolute Error)

$$ MAE = \frac{1}{n} \sum_{i=1}^{n} | y_i - \hat{y}_i | $$

- –†–µ–∞–ª—å–Ω–æ 1 –º–ª–Ω ‚Üí –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–æ 1.1 –º–ª–Ω (–æ—à–∏–±–∫–∞ 100 —Ç—ã—Å.)
- –†–µ–∞–ª—å–Ω–æ 2 –º–ª–Ω ‚Üí –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–æ 2.2 –º–ª–Ω (–æ—à–∏–±–∫–∞ 200 —Ç—ã—Å.)

 $$MAE = \frac{100000 + 200000 + 300000}{3}$$

## üéØ 2. MSE (–°—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–∞—è –æ—à–∏–±–∫–∞, Mean Squared Error)

$$ MSE = \frac{1}{n} \sum_{i=1}^{n} ( y_i - \hat{y}_i )^2 $$

‚úÖ –•–æ—Ä–æ—à–æ: —Å–∏–ª—å–Ω–µ–µ —à—Ç—Ä–∞—Ñ—É–µ—Ç –±–æ–ª—å—à–∏–µ –æ—à–∏–±–∫–∏, –ø–æ—ç—Ç–æ–º—É –º–æ–¥–µ–ª—å –±—É–¥–µ—Ç —Å—Ç–∞—Ä–∞—Ç—å—Å—è –∏—Ö –∏–∑–±–µ–≥–∞—Ç—å.

## 3. üéØ R¬≤ (–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –¥–µ—Ç–µ—Ä–º–∏–Ω–∞—Ü–∏–∏, –Ω–∞—Å–∫–æ–ª—å–∫–æ —Ö–æ—Ä–æ—à–æ –º–æ–¥–µ–ª—å –æ–±—ä—è—Å–Ω—è–µ—Ç –¥–∞–Ω–Ω—ã–µ)
